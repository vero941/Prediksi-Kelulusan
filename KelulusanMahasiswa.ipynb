```python
A = gabungan.iloc[:,2 :15]
```

```python
A 
```

```python
B = gabungan.iloc[:, 14]
```

```python
B
```

```python
from sklearn.model_selection import train_test_split
```

```python
A_train, A_test, B_train, B_test = train_test_split(A, B, test_size = 0.20, random_state = 36)
```

```python
from sklearn.tree import DecisionTreeClassifier as brn
```

```python
dtree = brn()
```

```python
dtree.fit(A_train, B_train)
```

```python
A_pred = dtree.predict(A_test)
```

```python
A_pred
```

```python
from sklearn.metrics import accuracy_score as acc
```

```python
acc = accuracy_score(B_test, A_pred)
```

```python
acc
```
